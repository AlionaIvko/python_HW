### Питання 5: 
Які 5 найбільш популярні бренди по Net Bill Amount в наших магазинах в розрізі міст?
    Побудуємо візуалізацію для більш зручного сприйняття.

```python

brand_sales_city = df.groupby(['Brand ', 'City'])['Net Bill Amount Eval'].sum().reset_index()


top_5_brands = brand_sales_city.groupby('Brand ')['Net Bill Amount Eval'].sum().nlargest(5).index
brand_sales_city_top5 = brand_sales_city[brand_sales_city['Brand '].isin(top_5_brands)]


plt.figure(figsize=(12, 8))
sns.barplot(data=brand_sales_city_top5, x='Brand ', y='Net Bill Amount Eval', hue='City', palette='Set2')
plt.title('Brand-wise Sales Distribution by City (Top 5 Brands)')
plt.xlabel('Brand ')
plt.ylabel('Net Bill Amount Eval')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title='City')
plt.show()
```

# Statistical hypothesis testing

### Питанння 10: 
Чи впливає пропозиція знижок по певним способам оплати на поведінку клієнтів?
Гіпотеза 0 - Пропозиція знижок по певним способам оплати немає впливу на поведінку клієнтів.
Гіпотеза 1 - Пропозиція знижок по певним способам оплати стимулює клієнтів купувати більше.

```python
# створюємо DataFrame для тестування
sales_data = {
    'Receipt Number': range(1, 101),
    'Payment Mode': np.random.choice(['Cash', 'Wallet'], size=100),
    'DISC': np.random.choice([0.01, 0.05], size=100), 
    'Net Bill Amount': np.random.uniform(30, 200, size=100)
}

df_sales = pd.DataFrame(sales_data)
```

```python
# визначаємо групи
groups = np.random.choice(['A', 'B'], size=len(df_sales))
df_sales['Group'] = groups
```

```python
# Застосуємо знижки залежно від групи та способу оплати
df_sales.loc[(df_sales['Group'] == 'A') & (df_sales['Payment Mode'] == 'Cash'), 'Net Bill Amount'] *= (1 - df_sales['DISC'])
df_sales.loc[(df_sales['Group'] == 'B') & (df_sales['Payment Mode'] == 'Wallet'), 'Net Bill Amount'] *= (1 - df_sales['DISC'])
```

```python
# Обчислимо Net Bill Amount для кожної групи
total_sales_net_bill = df_sales.groupby('Group')['Net Bill Amount'].sum()
```

```python
# Z-test
group_a_sales = df_sales[df_sales['Group'] == 'A']['Net Bill Amount']
group_b_sales = df_sales[df_sales['Group'] == 'B']['Net Bill Amount']
z_stat, p_value = stats.ttest_ind(group_a_sales, group_b_sales)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

#### Висновок 10: 
Оскільки 𝑝-значення=0.448 набагато вище за 𝛼= 0.05, ми не можемо відхилити нульову гіпотезу. 
    Результат означє  що пропозиція знижок по певним способам оплати немає впливу на поведінку клієнтів.
